# 3. Loading data from all stations and all years

We can load data from several stations and years in one go. The code below demonstrates how this can be done. If you have a lot of data, this can take a while.

If you have excel sheets with your own discharge data, you can adapt the library object below to read it in. Make sure that the data is in the same format as the data in the example data set or there will be errors when you try to read the data. For example, there must not be any notes in other columns than the first (which contains dates) and the second (which contains discharge values).

```python
# We can read data from several stations if we want to.
# Edit the station IDs or names here and the paths to the corresponding data files.
library = pd.DataFrame(
        {
            # The name of the stations
            "station": ["Sihl", "Rosegbach"],
            # The path to the files with daily river runoff data
            "file": ["D:/Articulo_SWAT/Sihl_example_river_runoff.xlsx", 
                     "D:/Articulo_SWAT/Rosegbach_example_river_runoff.xlsx"],
        }
    )

# Initiate the dictionary for the data to be read from excel
data_dict = {}

# Create a vector with numbers from 2000 to 2023, that is the years for 
# which the data is available in the excel sheets.    
years = np.arange(2000, 2003, 1)

# Get the file names for the stations
for index, row in library.iterrows():
    file_path = row["file"]
    station = row["station"]

    # Read the data for the station
    for year in years:
        df = pd.read_excel(
            file_path, sheet_name=str(year), header=[0], 
            skiprows=[1],  # Skip the header row that might contain different strings in different sheets
            names=['Date', 'Q_m3s'],  # Reassign column names
            parse_dates=['Date']  # Parse the dates in the Date column
        )
        # Sort df by Date
        df.sort_values(by=['Date'], inplace=True)

        datetime_column = df.iloc[:, 0]  # Dates are in the first column
        discharge_column = df.iloc[:, 1]  # Discharges are in the second column

        data = pd.DataFrame({
            "Date": datetime_column.values,
            "Q_m3s": discharge_column.values,
            "Year": year,
            "Code": station,
        })

        data_dict[station, year] = data


# Combine all sheets into a single DataFrame
combined_data = pd.concat(data_dict.values(), ignore_index=True)
    
# Make sure all Dates in the date column are really datetimes
combined_data.loc[:,'Date'] = pd.to_datetime(combined_data.loc[:,'Date'])

# Replace missing values with NaN
combined_data.replace(to_replace='-', value=np.nan, inplace=True)

combined_data
```

# 4. Plot the time series

We can also create time series plots for several stations in one go. The code below demonstrates how this can be done.

```python
# Plot the combined data Q_m3s over Date using holoviews, loop over all stations.
curves = {code: hv.Curve(data, 'Date', 'Q_m3s').opts(title=code) for code, data in combined_data.groupby('Code')}
# Put all curve plots in a single layout
layout = hv.Layout(list(curves.values())).cols(1)
layout.opts(opts.Curve(width=800, height=400, show_grid=True, framewise=True, 
                       tools=['hover'], xlabel='Date', ylabel='Discharge [m3/s]'))
layout	
```

# 5. Calculate and plot the mean hydrograph

The example data set only contains 3-4 years of data for each station. Not enough to calculate an average discharge. But we can still demonstrate how this can be done. The code below calculates the mean discharge for each day of the year and plots it.

```python
# Drop the 29th of February from the data
clean_data = combined_data[~((combined_data['Date'].dt.month == 2) & (combined_data['Date'].dt.day == 29))].copy()

# Add a column for the day of the year
clean_data.loc[:,'Day'] = clean_data.loc[:,'Date'].dt.dayofyear

# Get rid of the year and date columns
clean_data.drop(['Year', 'Date'], axis=1, inplace=True)

# Group by day of the year and calculate the mean discharge for each day
mean_hydrograph = clean_data.groupby(['Code', 'Day']).mean()

# Plot the hydrograph using holoviews
curves = {code: hv.Curve(data, 'Date', 'Q_m3s').opts(title=code) for code, data in mean_hydrograph.groupby('Code')}
layout = hv.Layout(list(curves.values())).cols(1)
layout.opts(opts.Curve(width=800, height=400, show_grid=True, framewise=True, 
                       tools=['hover'], xlabel='Day of the year', ylabel='Discharge [m3/s]'))
layout	
```

You might want to add the minimum and maximum discharge observed on each day or shaded areas for the 50- and 90-percentile ranges. The codes snipped below shows how this can be done.

```python
# Group by day of the year and calculate the mean discharge for each day
min_hydrograph = clean_data.groupby(['Code', 'Day']).min()
max_hydrograph = clean_data.groupby(['Code', 'Day']).max()
percentile05_hydrograph = clean_data.groupby(['Code', 'Day']).quantile(0.05)
percentile95_hydrograph = clean_data.groupby(['Code', 'Day']).quantile(0.95)
percentile25_hydrograph = clean_data.groupby(['Code', 'Day']).quantile(0.25)
percentile75_hydrograph = clean_data.groupby(['Code', 'Day']).quantile(0.75)

hydrograph_norm_perc = pd.concat([
    mean_hydrograph["Q_m3s"],
    min_hydrograph["Q_m3s"],
    max_hydrograph["Q_m3s"],
    percentile05_hydrograph["Q_m3s"],
    percentile25_hydrograph["Q_m3s"],
    percentile75_hydrograph["Q_m3s"],
    percentile95_hydrograph["Q_m3s"],
], axis=1)
hydrograph_norm_perc.columns = ["Mean", "Min", "Max", "Perc_05", "Perc_25", "Perc_75", "Perc_95"]	

curves = {code: hv.Area(data, kdims=["Day"], vdims=["Perc_05", "Perc_95"], 
        label="90- ") \
            .opts(alpha=0.2, color="#307096", line_width=0)*\
hv.Area(data, kdims=["Day"], vdims=["Perc_25", "Perc_75"],
        label="50- ") \
            .opts(alpha=0.4, color="#307096", line_width=0)*\
hv.Curve(data, 'Day', 'Mean', label='Mean').opts(width=800, height=400).\
    opts(title=code, xlabel='Day of the year', ylabel='Discharge [m3/s]', color="blue") * \
hv.Curve(data, 'Day', 'Min', label='Min').opts(width=800, height=400).\
    opts(alpha=0.3, color="green") * \
hv.Curve(data, 'Day', 'Max', label='Max').opts(width=800, height=400).\
    opts(alpha=0.3, color="red") for code, data in hydrograph_norm_perc.groupby('Code')}
layout = hv.Layout(list(curves.values())).cols(1)
layout
```
